```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('assets/sus.png')
```

```python
cv2.imshow('sus',img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
```
